## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #import matplotlib library and assigning as plt for easy use.
%matplotlib inline
```

## Basic plotting

```python
x = np.linspace(0,10,100) #creates 100 equal spaced value between 0 to 10
y = np.sin(x)
z = np.cos(x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,1,1]) #creates graph size with left,bottom,width,height
ax.plot(x,y, c = 'red', alpha = 0.8, marker ='^', lw = 2.5,label = "Sine") #sinusodial graph
ax.plot(x,z, c = 'green', alpha = 0.8, marker ='o', lw = 1.0, label = "Cosine") 
ax.set_xlim(2,10) #our graph x-axis limit is shrinken btwn 2 to 10
ax.legend()
ax.set_xlabel("Time") #name of x-axis
ax.set_ylabel("Amplitude")
ax.set_title("Matplotlib plot")

#alpha is used to make transparency
#c is used for color argument
#marker is used for displaying points
#lw - linewidth which is used for thickness of line
#label - displays the name of the graph
```

## SubPlots

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (10,6)) #produces 1 X 2 graph
#figsize helps us to increase the height and width of the graph

ax[0].plot(x,y, c= 'red', marker = 'o')
ax[1].plot(x,z, c ='green', marker = 'x')
```
