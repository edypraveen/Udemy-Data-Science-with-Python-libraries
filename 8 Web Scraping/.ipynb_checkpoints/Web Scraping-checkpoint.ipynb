## Web Scraping
    
- Web Scraping Process
- Extract title,link,images, individual tag from webpages
- Scrap IMDB movie listing page
- Practical Activity
    

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.imdb.com/chart/boxoffice/'
html = requests.get(url)
```

```python
bsobj = BeautifulSoup(html.text, 'lxml')
```

```python
bsobj
```

```python
bsobj.find('title')
```

```python
bsobj.find('title').getText() #display oly the title name without html codes
```

```python
h1List = bsobj.findAll('h1')
```

```python
h1List
```

```python
for h1 in h1List:
    print(h1.getText())
```

```python
thList = bsobj.findAll('th') #7 times
```

```python
for th in thList:
    print(th.getText())
```

```python
#Grab the titles of movie List
```

```python
titleList = bsobj.findAll('td',{'class':'titleColumn'})
```

```python
for title in titleList:
    print(title.getText())
```
