```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
df = pd.DataFrame(data = np.random.randint(1,100,(100,3)), columns=["C1",'C2','C3']) 
#generates random value from 1 to 100 in 100X3
```

```python
df.shape #displays rows,columns
```

```python
df.head()
```

## Area Plot

```python
df['C2'].plot.area() #in panda, we must introduce data to be plotted in advance. here data to be plotted is C2 column
```

```python
df.plot.area() #for all the values
```

## Bar Plot

```python
df['C3'].plot.bar() #but as it too much data, the visualization becomes clumsy
```

```python
df['C3'].plot.barh() #in horizontal view
```

## Line Plot

```python
df['C1'].plot.line()
```

## Box Plot

```python
df.plot.box() #displays mean,min value, max value
```

## Kernel Density

```python
df['C1'].plot.kde()
```

## Scatterplot

```python
df.plot.scatter(x = 'C1', y = 'C2')
```

## Histogram

```python
df['C3'].plot.hist(bins = 20) #data are equally spaced to 20 (0-5,5-10,10-15..95-100)
```
